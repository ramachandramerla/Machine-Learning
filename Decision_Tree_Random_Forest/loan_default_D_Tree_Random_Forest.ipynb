```python
#create a dummies for this column  
cat_feats = ['purpose']
```

```python
#this will return data type as boolean for defalut but we can use dtype to modify the default
final_data = pd.get_dummies(data=loans,columns=cat_feats,drop_first=True,dtype=int)
```

```python
final_data.head()
```

```python
final_data.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

```python
pred_dtc = dtc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,pred_dtc))
print(confusion_matrix(y_test,pred_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train,y_train)
```

```python
pred_rfc = rfc.predict(X_test)
```

```python
print(classification_report(y_test,pred_rfc))
print(confusion_matrix(y_test,pred_rfc))
```
