```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2)
```

```python
df.columns
```

```python
model.fit(df.drop('Private',axis=1))
```

```python
model.cluster_centers_
```

```python
model.inertia_
```

```python
#creating a function to give 1 for yes and 0 for no
def convert(private):
    if private == 'Yes':
        return 1
    else:
        return 0
```

```python
#creating a column cluster and calling a function for which column needs to apply the condition
df['Cluster'] = df['Private'].apply(convert)
```

```python
df.head(10)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

for unsupervised learning we dont need to split the data

```python
print(confusion_matrix(df['Cluster'],model.labels_))
print(classification_report(df['Cluster'],model.labels_))
```
