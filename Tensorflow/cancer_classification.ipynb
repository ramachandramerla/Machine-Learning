Model

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=101,test_size=0.25)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

#for binary classifcation we need to add sigmoid function for the last nueron
model.add(Dense(1,activation='sigmoid'))

model.compile(loss = 'binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,
          validation_data = (X_test,y_test),
          epochs = 600)
```

```python
loss_data = pd.DataFrame(model.history.history)
```

```python

loss_data.plot()
```

How to stop the Overfitting

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

#for binary classifcation we need to add sigmoid function for the last nueron
model.add(Dense(1,activation='sigmoid'))

model.compile(loss = 'binary_crossentropy',optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',
                           mode='min',
                           verbose=1,
                           patience=25)
```

```python
model.fit(x=X_train,y=y_train,
          validation_data = (X_test,y_test),
          epochs = 600,callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))
#for binary classifcation we need to add sigmoid function for the last nueron
model.add(Dense(1,activation='sigmoid'))

model.compile(loss = 'binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,
          validation_data = (X_test,y_test),
          epochs = 600,callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
predicitions  = (model.predict(X_test)>0.5)*1
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predicitions))
print(confusion_matrix(y_test,predicitions))
```
