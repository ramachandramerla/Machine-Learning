```python
#no need of id column
df = df.drop('id',axis=1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
#date column from object to datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].apply(lambda date : date.year)
```

```python
df['Month'] = df['date'].apply(lambda date : date.month)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Month',y='price')
```

```python
#group by month to understand better than the box plot plotted above
df.groupby('Month').mean()['price'].sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,5))
df.groupby('Month').mean()['price'].plot()
```

```python
df.groupby('year').mean()['price'].sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,5))
df.groupby('year').mean()['price'].plot()
```
