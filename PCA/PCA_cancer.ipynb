```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
type(cancer)
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
cancer.target
```

```python
cancer.target_names
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
#converting the 30 columns
scaled_data = scaler.transform(df)
```

```python
scaled_data.shape
```

```python
from sklearn.decomposition import PCA
```

```python
#converting the 30 columns to 2 columns
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
x_pca.shape
```

```python
#plot a scatterplot between 2 columns
#color use "c"

plt.scatter(x=x_pca[:,0],y=x_pca[:,1],c=cancer['target'])
plt.xlabel("1st column")
plt.ylabel("2nd column")
plt.title("plot btw 2 columns which are converted")
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(data = pca.components_,columns=cancer['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(10,4))
sns.heatmap(df_comp)
```
